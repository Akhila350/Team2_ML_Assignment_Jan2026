 # Hypothesis Statements

# Hypothesis 1: Education and Income

 Null Hypothesis (H‚ÇÄ): Education level has no significant relationship with income level.

 Alternative Hypothesis (H‚ÇÅ): Individuals with higher education levels are more likely to earn more than $50K.

 Variables used: education-num, income

 Test Used: Chi-Square Test (Categorical vs Categorical)

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create contingency table: Education vs Income
contingency = pd.crosstab(df['education'], df['income'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency)

# Print results
print("Chi-square statistic:", chi2)
print("p-value:", p)

# Interpret result
if p < 0.05:
    print("Reject the null hypothesis: Education level affects income.")
else:
    print("Fail to reject the null: No significant relationship between education and income.")
```

<p style="font-family: Cambria; font-size: 22px;"><b> Hypothesis 2: Age and Income

> Null Hypothesis (H‚ÇÄ): Age has no significant effect on income level.

> Alternative Hypothesis (H‚ÇÅ): Individuals in higher age groups are more likely to earn more than $50K.

>> Variables used: age, income

> Test Used: Independent T-Test (Numerical vs Binary)

```python
from scipy.stats import ttest_ind

# Separate age values based on income group
age_low_income = df[df['income'] == '<=50K']['age']
age_high_income = df[df['income'] == '>50K']['age']

# Run independent t-test
t_stat, p_value = ttest_ind(age_low_income, age_high_income, equal_var=False)

# Output the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: Age is significantly different between income groups.")
else:
    print("Fail to reject the null: No significant age difference between income groups.")
```

```python
# Define categorical columns to test
categorical_cols = [
    'workclass',
    'education',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native-country'
]

# Loop through each categorical column
for col in categorical_cols:
    # Create contingency table
    contingency = pd.crosstab(df[col], df['income'])

    # Perform Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency)

    # Output
    print(f"\nüìä Column: {col}")
    print(f"Chi2 Statistic: {chi2:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print(f"P-value: {p:.4f}")

    if p < 0.05:
        print("‚úÖ Statistically significant association with income.")
    else:
        print("‚ùå No significant association with income.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind


# Split the data by income group
edu_low_income = df[df['income'] == '<=50K']['education-num']
edu_high_income = df[df['income'] == '>50K']['education-num']

# Calculate the means
mean_low = edu_low_income.mean()
mean_high = edu_high_income.mean()

# Run Welch's t-test
t_stat, p_val = ttest_ind(edu_low_income, edu_high_income, equal_var=False)

# Print mean values
print(f"Mean education-num (<=50K): {mean_low:.2f}")
print(f"Mean education-num (>50K): {mean_high:.2f}\n")

# Print test result
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpret result
if p_val < 0.05:
    print("‚úÖ Reject the null hypothesis: Education level differs significantly between income groups.")
else:
    print("‚ùå Fail to reject the null hypothesis: No significant difference in education level between income groups.")
```

```python
# List of numeric columns to analyze
numeric_cols = ['age','hours-per-week','education-num','fnlwgt']


# Loop through each column and perform t-test
for col in numeric_cols:
    group_low = df[df['income'] == '<=50K'][col]
    group_high = df[df['income'] == '>50K'][col]
    
    # Calculate means
    mean_low = group_low.mean()
    mean_high = group_high.mean()

    # Perform Welch‚Äôs t-test
    t_stat, p_val = ttest_ind(group_low, group_high, equal_var=False)

    # Output
    print(f"\nüìä Column: {col}")
    print(f"Mean for <=50K: {mean_low:.2f}")
    print(f"Mean for >50K: {mean_high:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_val:.4f}")

    if p_val < 0.05:
        print("‚úÖ Statistically significant difference between income groups.")
    else:
        print("‚ùå No statistically significant difference between income groups.")
```

# Mann‚ÄìWhitney U tests (Skewed numerical)

```python
from scipy.stats import mannwhitneyu

# List of skewed numeric columns
skewed_cols = ['capital-gain', 'capital-loss']

# Loop through each column and perform Mann‚ÄìWhitney U test
for col in skewed_cols:
    group_low = df[df['income'] == '<=50K'][col].dropna()
    group_high = df[df['income'] == '>50K'][col].dropna()

    # Calculate medians
    median_low = group_low.median()
    median_high = group_high.median()

    # Perform Mann‚ÄìWhitney U test
    u_stat, p_val = mannwhitneyu(group_low, group_high, alternative='two-sided')

    # Output
    print(f"\nüìä Column: {col}")
    print(f"Median for <=50K: {median_low:.2f}")
    print(f"Median for >50K: {median_high:.2f}")
    print(f"U-statistic: {u_stat:.4f}")
    print(f"P-value: {p_val:.4f}")

    if p_val < 0.05:
        print("‚úÖ Statistically significant difference between income groups.")
    else:
        print("‚ùå No statistically significant difference between income groups.")
```

# Hypothesis Testing
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. The goal is to test an assumption (hypothesis) about a population parameter and determine whether the evidence supports or refutes the hypothesis.
Steps in Hypothesis Testing:
Formulate Hypotheses:
1. Null Hypothesis (H‚ÇÄ):
 The null hypothesis is the statement or claim being made (which we are trying to disprove).
H‚ÇÄ: income <=50K
2. Alternative Hypothesis (H‚ÇÅ or Ha):
The alternative hypothesis is the hypothesis that we are trying to prove, and which is accepted if we have sufficient evidence to reject the null hypothesis.
H‚ÇÅ: income >50K
Choose a Significance Level (Œ±):
This is the probability threshold below which the null hypothesis will be rejected. Common significance levels are 0.05 (5%) or 0.01 (1%).
If Œ± = 0.05, there's a 5% risk of rejecting the null hypothesis when it is true (Type I error).
Select the Appropriate Test:
Depending on the data and the hypothesis, various statistical tests can be used (e.g., t-test, chi-square test, ANOVA). Each test has assumptions about the data that need to be met.
Compute the Test Statistic and P-value:
The test statistic (e.g., z-score, t-score) quantifies the difference between the sample data and what is expected under the null hypothesis.
The p-value is the probability of obtaining a result as extreme as, or more extreme than, the observed result if the null hypothesis is true.
Make a Decision:
Reject H‚ÇÄ if the p-value is less than the significance level (p-value < Œ±). This suggests the data provides enough evidence to support the alternative hypothesis.
Fail to reject H‚ÇÄ if the p-value is greater than the significance level (p-value ‚â• Œ±). This suggests the data is not strong enough to reject the null hypothesis.

# Type 1 and Type 2 errors

Type I error (Œ±): rejecting a true H‚ÇÄ (false positive)

Type II error (Œ≤): failing to reject a false H‚ÇÄ (false negative)

Power (1 ‚àí Œ≤): ability to detect a true effect

minimize Type II error by selecting the correct statistical test and sufficient sample size.
